```python
import pandas as pd # Import the Pandas library that will allow us to work with data in a tabular format
import matplotlib.pyplot as plt # Import the Matplotlib library that will allow us to create data visualizations
from sklearn.cluster import KMeans # Import the KMeans class from the sc

# Load the CSV file
df = pd.read_csv('simple.csv')

# Display the first few rows
print(df.head(10))

# Select relevant features for clustering
X = df[['Age', 'WhiteBloodCellCount', 'RedBloodCellCount', 'PlateletCount']]

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Add cluster labels to the DataFrame
df['Cluster'] = kmeans.labels_

# Display the updated DataFrame
print(df)

# Visualize the clusters using Age and WhiteBloodCellCount
plt.scatter(df['Age'], df['WhiteBloodCellCount'], c=df['Cluster'], cmap='viridis', marker='o', edgecolor='k', s=100)
plt.title('Patient Clusters')
plt.xlabel('Age')
plt.ylabel('White Blood Cell Count')
plt.colorbar(label='Cluster')
plt.show()
```
